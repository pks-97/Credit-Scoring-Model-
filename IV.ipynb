```python
import numpy as np
import pandas as pd
def iv(df, feature, target, minimum, maximum, step):
    lst = []
    i = minimum

    while i < maximum:
        lst.append([feature,
                    i,
                    df[((df[feature]).astype('float') >= i) & ((df[feature]).astype('float') < i + step)].count()[feature],
                    df[((df[feature]).astype('float') >= i) & ((df[feature]).astype('float') < i + step) & ((df[target]) == 0)].count()[feature],
                    df[((df[feature]).astype('float') >= i) & ((df[target]) == 1) & ((df[feature]).astype('float') < i + step)].count()[feature]])
        i = i + step
    df[feature] = df[feature].fillna("NULL")
    lst.append([feature,
                    "NULL",
                    df[((df[feature]) == "NULL")].count()[feature],
                    df[((df[feature]) == "NULL") & ((df[target]) == 0)].count()[feature],
                    df[((df[feature]) == "NULL") & ((df[target]) == 1)].count()[feature]])
    d_yoj = pd.DataFrame(lst, columns=['Variable', 'Value', 'All', 'Good', 'Bad'])
    d_yoj['Share'] = d_yoj['All'] / d_yoj['All'].sum()
    d_yoj['Bad Rate'] = d_yoj['Bad'] / d_yoj['All']
    d_yoj['Distribution Good'] = (d_yoj['All'] - d_yoj['Bad']) / (d_yoj['All'].sum() - d_yoj['Bad'].sum())
    d_yoj['Distribution Bad'] = d_yoj['Bad'] / d_yoj['Bad'].sum()
    d_yoj['WoE'] = np.log(d_yoj['Distribution Good'] / d_yoj['Distribution Bad'])

    d_yoj = d_yoj.replace({'WoE': {np.inf: 0, -np.inf: 0}})

    d_yoj['IV'] = d_yoj['WoE'] * (d_yoj['Distribution Good'] - d_yoj['Distribution Bad'])

    d_yoj = d_yoj.sort_values(by=['Variable', 'Value'], ascending=[True, True])
    d_yoj.index = range(len(d_yoj.index))


    iv_yoj= d_yoj['IV'].sum()

    return iv_yoj, d_yoj
```

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
import numpy as np
```

```python
d = pd.read_csv("good.csv")
```

```python
y = d.iloc[:,1]
```

```python
#YOJ,CLNO,REASON and MORTDUE should be eliminated
final = d.drop(['CLNO', 'YOJ','REASON','MORTDUE'], axis=1)
```

```python
x = final.iloc[:,2:]
```

```python
x
```

```python
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.33, random_state=1)
```

```python
X_train
```

```python
y_train
```

```python
classifier = LogisticRegression(tol = 0.0000000000000000000000000000000000000000001,max_iter=1000)
classifier.fit(X_train, y_train)
```

```python
y_pred = classifier.predict(X_test)
```

```python
print('Accuracy of logistic regression classifier on test set: {:.2f}'.format(classifier.score(X_test, y_test)))
```

```python
from sklearn.svm import SVC
```

```python
clf = SVC(kernel='linear')
clf.fit(X_train, y_train)
```

```python
a = clf.predict(X_test)
```
